```python
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer(
    "codefuse-ai/F2LLM-1.7B",
    trust_remote_code=True
)
```

```python
from bertopic import BERTopic

topic_model = BERTopic.load(
    "cfpb_bertopic_model",
    embedding_model=embedding_model
)
```

```python
import pandas as pd

lookup = pd.read_csv("/content/drive/MyDrive/BERTopic_Project/data/processed/topic_lookup.csv")

topic_lookup = (
    lookup
    .set_index("Topic")["Topic_Label"]
    .to_dict()
)
```

```python
def predict_topic(text):
    topics, probs = topic_model.transform([text])

    topic_id = topics[0]
    topic_name = topic_lookup.get(topic_id, "Unknown Topic")

    print("=" * 70)
    print("INPUT:")
    print(text)
    print("=" * 70)
    print(f"Topic ID    : {topic_id}")
    print(f"Topic Name  : {topic_name}")
    print(f"Probability : {probs[0]:.3f}")
```

```python
predict_topic("Someone stole my credit card")
```

```python
predict_topic("Identity Theft Fraud Reporting")
```
